```python
import os 
import pandas as pd
from dotenv import load_dotenv
from sqlalchemy import create_engine

# --- Connect to DB
load_dotenv('.env')
DB_CONNECTION_URL = os.getenv("POSTGRES_URL")
print(DB_CONNECTION_URL)

engine = create_engine(DB_CONNECTION_URL)


SCHEMA_CREATION_QUERY = 'CREATE SCHEMA "public";'


TABLE_CREATION_QUERY = """CREATE TABLE IF NOT EXISTS "currency_quotes_bronze" (
	"id" bigint PRIMARY KEY GENERATED ALWAYS AS IDENTITY (sequence name "currency_quotes_bronze_id_seq" INCREMENT BY 1 MINVALUE 1 MAXVALUE 9223372036854775807 START WITH 1 CACHE 1),
	"quote_date" timestamp,
	"currency_code" varchar(10),
	"type" varchar(5),
	"currency" varchar(5),
	"buy_rate" numeric(20, 10),
	"sell_rate" numeric(20, 10),
	"parity_buy" numeric(20, 10),
	"parity_sell" numeric(20, 10),
	"processing_date" timestamp DEFAULT CURRENT_TIMESTAMP
);"""

CREATE_INDEX_QUERY = 'CREATE UNIQUE INDEX "currency_quotes_bronze_pkey" ON "currency_quotes_bronze" ("id");'

SIMPLE_SELECT_QUERY = "SELECT * FROM currency_quotes_bronze"
conn = engine.connect()
try: 
    data = conn.exec_driver_sql(SIMPLE_SELECT_QUERY)
    print('Sucess')
    df = pd.DataFrame(data.all(), columns=data.keys())
    print(df)
    # conn.commit()
except Exception as e:
    print(e)
finally:
    conn.close()
```
